#### Importing the required libraries. Here numpy is used to convert the data into the numerical format after it is read from the DAT file. Matplotlib is the plotting library for python used for Data Visualization and Scipy is being used for finding the optimal weights and bias for higher order functions.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

#### Reading the file using the open method that built-in method of python. The file is read line by line and then each line is split to get the X and y values. Finally, the list of values are converted to the numpy arrays.

```python
f = open('hw2.dat', 'r')
arr = f.readlines()

x, y = [], []
for i in arr:
    a, b = i.split()
    x.append(a)
    y.append(b)

x, y = np.array(x, dtype='float64'), np.array(y, dtype='float64')
```

#### Defining the functions for evaluating the model including mean squared error, root mean squared error and r-squared error and for calculating the least squares.

```python
def mse(true, pred):
    return np.mean((pred-true)**2)

def rmse(true, pred):
    return mse(true, pred)**0.5

def r_squared(true, pred):
    true_mean = true.mean()
    pred_mean = pred.mean()
    tot = ((true-true_mean)**2).sum(axis=0)
    obs = ((true-pred)**2).sum(axis=0)
    return 1-(obs/tot)

def least_squares(x, y):
    xmean = x.mean()
    ymean = y.mean()
    num = ((x-xmean)*(y-ymean)).sum(axis=0)
    den = ((x-x.mean())**2).sum(axis=0)
    weight = num / den
    bias = ymean - (weight*xmean)
    
    return weight, bias
```

### Degree = 1

#### Here, the model is fitted using the independent variable and the co-efficient(s) and bias. Then, the evaluation measures for the model of the specific degree are calculated.

```python
def predict(x, weight, bias):
    return (weight * x) + bias
```

```python
weight, bias = least_squares(x, y)
print('Coefficient :{} , Intercept : {}'.format(weight, bias))
ypred1 = predict(x, weight, bias)
ytrue = y
```

```python
print('Mean Squared Error : ', mse(ytrue, ypred1))
print("Root Mean Squared Error : ", rmse(ytrue, ypred1))
print("R-Squared Value: ", r_squared(ytrue, ypred1))
```

### Degree = 3

```python
def predict(x, weight1, weight2, weight3, bias):
    return (weight1 * x) + (weight2 * x**2) + (weight3 * x**3) + bias

popt, _ = curve_fit(predict, x, y)
weight1, weight2, weight3, bias = popt
```

```python
print('Coefficient1 :{} , Coefficient2 :{} , Coefficient3 :{} , Intercept : {}'.format(weight1, weight2, weight3, bias))
ypred2 = predict(x, weight1, weight2, weight3, bias)
ytrue = y
```

```python
print('Mean Squared Error : ', mse(ytrue, ypred2))
print("Root Mean Squared Error : ", rmse(ytrue, ypred2))
print("R-Squared Value: ", r_squared(ytrue, ypred2))
```

```python
print('Coefficient1 :{} , Coefficient2 :{} , Coefficient3 :{} , Coefficient4 :{} , Coefficient5 :{} , Intercept : {}'.format(weight1, weight2, weight3, weight4, weight5, bias))
ypred3 = predict(x, weight1, weight2, weight3, weight4, weight5, bias)
ytrue = y
```
