```python
training_set.shape
```

#### Dropping any features with pearson correlation greater than 0.22

```python
corr_matrix = training_set.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

to_drop = [column for column in upper.columns if any(upper[column] > 0.22)]
to_drop
```

```python
training_set = training_set.drop(to_drop, axis=1)
training_set.shape
```

```python
col_names = training_set.columns
col_names
```

```python
len(col_names)
```

```python
for name in col_names:
    training_set[name] = training_set[name].fillna(training_set[name].mean())
```

```python
training_set.isna().sum().any()
```

```python
training_set.head()
```

#### Scaling the feature values with MinMaxScaler()

```python
scaler.fit(training_set)
training_set = scaler.transform(training_set)
```
